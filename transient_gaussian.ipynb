```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pnts = 4000
srate = 1000
time = np.arange(0, pnts-1)/srate - 1

peaktime = 1
fwhm = 0.4
sin_f = 7
```

```python
gaus = np.exp(-(4*np.log(2) * (time-peaktime) ** 2) / fwhm ** 2)
cosw = np.cos(2*np.pi*sin_f*time + 2*np.pi*np.random.rand(1))
signal = cosw * gaus
```

```python
plt.plot(time, signal)
```
