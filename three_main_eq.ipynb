```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
f = 2
s_rate = 1000
time = np.arange(-1, 1, 1/s_rate)
amp = 2
phase = np.pi/3
```

```python
sine_wave = amp * np.sin(2 * np.pi * f * time + phase)
```

```python
plt.plot(time, sine_wave)
plt.xlabel('Time (ms)')
plt.title("sine wave")
```

```python
f = [3, 10, 5, 15, 35]
amp = [5, 15, 10, 5, 7]
phases = [np.pi/7, np.pi/2, np.pi/3, np.pi/4, np.pi/5]
sine_wave = np.zeros((len(f), len(time)))

for i in range(len(f)):
    sine_wave[i] = amp[i] * np.sin(2 * np.pi * f[i] * time + phases[i])
```

```python
fig, axs = plt.subplots(6, 1, figsize=(10, 10), sharex=True)

for i in range(len(f)):
    axs[i].plot(time, sine_wave[i])
    axs[i].set_title("sine wave: {} Hz".format(f[i]))

axs[5].set_title("sum of sine waves")
axs[5].plot(time, np.sum(sine_wave, axis=0))
```
