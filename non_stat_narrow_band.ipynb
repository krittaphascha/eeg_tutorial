```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_signal
```

```python
pnts = 4567
srate = 987

peakfreq = 14
fwhm = 5

hz = np.linspace(0,srate,pnts)
```

```python
s = fwhm * (2*np.pi-1) / (4*np.pi)
x = hz-peakfreq
fg = np.exp(-0.5 * (x/s)**2)
```

```python
fc = np.random.rand(1, pnts) * np.exp(1j * 2 * np.pi * np.random.rand(1, pnts))

fc = fc * fg

signal = 2*np.real(np.fft.ifft(fc))
```

```python
plot_signal(signal, sampling_rate=srate)
```
