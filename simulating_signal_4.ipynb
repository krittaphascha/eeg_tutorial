```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Pure phase-locked sine wave

```python
srate = 500
pnts = 1500
trials = 30
n_channels = 23
```

```python
sinefreq = 6.75
```

```python
times = np.arange(0, pnts) / srate
```

```python
data = np.zeros((n_channels, pnts, trials))
for i in range(n_channels):
    for j in range(trials):
        data[i, :, j] = np.sin(2* np.pi * sinefreq*times)
```

```python
plt.plot(data[10, :, :].mean(axis=-1))
plt.show()
```

## non phase lock sine wave

```python
data = np.zeros((n_channels, pnts, trials))
for i in range(n_channels):
    for j in range(trials):
        data[i, :, j] = np.sin(2 * np.pi *sinefreq * times + 2* np.pi* np.random.randn())

plt.plot(data[2, :, :].mean(axis=-1))
plt.show()
```

# multisine wave

```python
sine_freqs = [3, 5, 16]
amps = [2, 4, 5]


data = np.zeros((n_channels, pnts, trials))
for i in range(n_channels):
    for j in range(trials):
        for k in range(len(sine_freqs)):
            data[i, :, j] = data[i,: ,j] + amps[k] * np.sin(2 * np.pi * sine_freqs[k] * times)

plt.plot(data[10, :, :].mean(axis=-1))
plt.show()
```

## Non stationary sine wave

```python
data = np.zeros((n_channels, pnts, trials))
for i in range(n_channels):
    for j in range(trials):
        freq_mod = 20 * np.interp(np.linspace(0, 1, pnts), np.linspace(0, 1, 100), np.random.randn(100))
        data[i, :, j] = np.sin(2 * np.pi *sinefreq * times + 2* np.pi* np.random.randn())

plt.plot(data[2, :, :].mean(axis=-1))
plt.show()
```
